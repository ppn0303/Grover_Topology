```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class result_obj():
    def __init__(self):
        self.epoch = []
        self.loss_train = []
        self.loss_val = []
        self.loss_val_av = []
        self.loss_val_bv = []
        self.loss_val_fg = []
        self.loss_val_topo = []
        self.loss_val_node = []

    def add_result(self, result):
        self.epoch.append(np.int(str(result).split(' ')[1]))
        self.loss_train.append(np.float(str(result).split(' ')[3]))
        self.loss_val.append(np.float(str(result).split(' ')[5]))
        self.loss_val_av.append(np.float(str(result).split(' ')[7]))
        self.loss_val_bv.append(np.float(str(result).split(' ')[9]))
        self.loss_val_fg.append(np.float(str(result).split(' ')[11]))
        self.loss_val_topo.append(np.float(str(result).split(' ')[13]))
        self.loss_val_node.append(np.float(str(result).split(' ')[15]))
    
    def plot_loss_val(self):
        plt.title("validation Loss")
        plt.plot(self.epoch, self.loss_val)
        
    def plot_loss_val_av(self):
        plt.title("validation Atom-view Loss")
        plt.plot(self.epoch, self.loss_val_av)
        
    def plot_loss_val_bv(self):
        plt.title("validation Bond-view Loss")
        plt.plot(self.epoch, self.loss_val_bv)
        
    def plot_loss_val_fg(self):
        plt.title("validation Motif Loss")
        plt.plot(self.epoch, self.loss_val_fg)

    def plot_loss_val_topo(self):
        plt.title("validation Topology Loss")
        plt.plot(self.epoch, self.loss_val_topo)
        
    def plot_loss_val_node(self):
        plt.title("validation Node Loss")
        plt.plot(self.epoch, self.loss_val_node)
```

```python
file = open('results.txt', 'r')
lines = file.readlines()
results = []
for line in lines:
    results.append(line)
file.close()
```

```python
res = result_obj()
```

```python
for i in range(len(results)):
    res.add_result(np.array(results[i]))
```

```python
res.plot_loss_val()
```

# text에서 epoch 해당만 찾기

```python
file = open('logs/merge_12.txt', 'r')
lines = file.readlines()
results = []
for line in lines:
    results.append(line)
file.close()
```

```python
epochs = []
for i in range(len(results)):
    if results[i].split(' ')[0] == "Epoch:":
        epochs.append(results[i])
```

```python
f = open("results.txt",'a')
for i in range(len(epochs)):
    data = epochs[i]
    f.write(data)
f.close()
```
